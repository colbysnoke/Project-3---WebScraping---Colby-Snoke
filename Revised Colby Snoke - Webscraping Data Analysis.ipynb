```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

# Which movie had the most days as the #1 release in 2021?

```python
#control u to bring up structure
url='https://www.boxofficemojo.com/daily/2021/?view=year'
response = requests.get(url)
status = response.status_code
if status == 200:
    page = response.text
    soup = bs(page)
else:
    print(f"Oops! Received status code {status}")
```

```python
movie_list = [ ]
stats_table = soup.find(id="table")
for row in stats_table.find_all('tr'):
    movie = row.find(class_ = "mojo-field-type-release").text
    movie_list.append({'Movie':movie})
movie_list
```

```python
movie_df = pd.DataFrame(movie_list)
movie_df
```

```python
movie_total = movie_df['Movie'].value_counts()
movie_total
```

# What Holiday had the highest cumulative gross in 2019?

```python
url='https://www.boxofficemojo.com/holiday/by-year/2019/?grossesOption=calendarGrosses'
response = requests.get(url)
status = response.status_code
if status == 200:
    page = response.text
    soup = bs(page)
else:
    print(f"Oops! Received status code {status}")
```

```python
movie_list2 = [ ]
stats_table = soup.find(id='table')
for row in stats_table.find_all('tr'):
    holiday = row.find(class_="mojo-field-type-named_interval").text
    gross = row.find(class_="mojo-field-type-money").text
    movie_list2.append({'holiday': holiday, 'Gross': gross})
movie_list2
```

```python
movie_df2 = pd.DataFrame(movie_list2)
movie_df2
```

```python
movie_df2 = movie_df2.drop([0])
```

```python
movie_df2['Gross'] = movie_df2['Gross'].replace('[\$,]', '', regex=True).astype(float)
```

```python
print(movie_df2[movie_df2['Gross'] == max(movie_df2['Gross'])].holiday)
```
